### 面积图

```python
from pyecharts.charts import Line
import pyecharts.options as opts

attr = ['衬衫','羊毛衫','雪纺衫','裤子','高跟鞋','袜子']
v1 = [5,20,36,10,75,90]
v2 = [10,25,8,60,20,80]
line= (
        Line()
       .add_xaxis(attr)
       .add_yaxis('商家A', v1, is_smooth=True,areastyle_opts=opts.AreaStyleOpts(opacity=0.3))
       .add_yaxis('商家B',v2, areastyle_opts=opts.AreaStyleOpts(opacity=0.8),
                  label_opts=opts.LabelOpts(is_show=False))
       .set_global_opts(title_opts=opts.TitleOpts(title="面积图"))
    )
line.render_notebook() 
#一个光滑，一个不光滑；一个有数值，一个没有，仅供展示
```

```python
from pyecharts.charts import Funnel
data = [("衬衫",20), ("羊毛衫",40), ("雪纺衫",60), ("裤子",80), ("高跟鞋",100), ("袜子",120)]
#funnel = Funnel()
funnel=(
    Funnel(init_opts=opts.InitOpts(width="1000px",height="600px"))
    .add(
        series_name="商品",
        data_pair=data,
        gap=2,
        tooltip_opts=opts.TooltipOpts(trigger="item",formatter="{a} <br/>{b}：{c}"),
        label_opts=opts.LabelOpts(is_show=True,position="inside"),
        itemstyle_opts=opts.ItemStyleOpts(border_color="#fff",border_width=1)
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="商品",subtitle="销量"))
)
funnel.render_notebook() 
```

### 雷达图

```python
from pyecharts import options as opts
from pyecharts.charts import Page, Radar

v1 = [[4300, 10000, 28000, 35000, 50000, 19000]]
v2 = [[5000, 14000, 28000, 31000, 42000, 21000]]

def radar_base() -> Radar:
    c = (
        Radar()
        .add_schema(
            schema=[
                opts.RadarIndicatorItem(name="KDA", max_=6500),
                opts.RadarIndicatorItem(name="输出", max_=16000),
                opts.RadarIndicatorItem(name="经济", max_=30000),
                opts.RadarIndicatorItem(name="生存", max_=38000),
                opts.RadarIndicatorItem(name="推进", max_=52000),
                opts.RadarIndicatorItem(name="打野", max_=25000),
            ]
        )
        .add("鲁班", v1,            #添加系列名称及数据
             color="red",           #设置边框线的颜色
             areastyle_opts = opts.AreaStyleOpts(#设置填充的属性
                 opacity = 0.5,                  #透明度
                 color="red"                     #填充颜色
         ),)
        .add("后裔", v2,color="blue",
             areastyle_opts = opts.AreaStyleOpts(
                 opacity = 0.5,#透明度
                 color="blue"
         ),)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title="英雄成长对比"))
    )
    return c
radar_base().render_notebook()
```

```python
import pandas as pd
import numpy as np
from pyecharts.charts import Line
import pyecharts.options as opts

path2='D:/python/郭峰Python讲义/数据与结果/07数据可视化/'
f=open(path2+'line_test.csv','r',encoding='utf8')
line_test= pd.read_csv(f,header=0,sep=',')
print(line_test.head())
month=list(line_test['month'])
download=list(line_test['download'])
browse=list(line_test['browse'])
```

```python
line = (
    Line()
    .add_xaxis(month)
    .add_yaxis("download", download,label_opts=opts.LabelOpts(is_show=True),symbol='circle',
               linestyle_opts=opts.LineStyleOpts(color="red",width=1.5))
    .add_yaxis("browse", browse,label_opts=opts.LabelOpts(is_show=False),symbol ='diamond',
               linestyle_opts=opts.LineStyleOpts(color="blue",width=1.5,type_='dotted')
                 
))

line.set_global_opts(title_opts=opts.TitleOpts(title="Download and Browse in Each Month", subtitle="副标题"),
                    legend_opts=opts.LegendOpts(is_show=False),
                    xaxis_opts=opts.AxisOpts(name="Month",is_show=True,
                                            name_location='middle',name_gap=30),
                    yaxis_opts=opts.AxisOpts(name="Number of Papers",is_show=True,
                                            name_location='middle',name_gap=60))

line.render_notebook() 
#本图一个显示具体数值，一个不显示，仅为演示
```

<font face="微软雅黑" size=3>实战2：回归系数图</font>  

```python
#画图
import pandas as pd
import numpy as np
from pyecharts.charts import Line
import pyecharts.options as opts

path2='D:/python/郭峰Python讲义/数据与结果/07数据可视化/'
f=open(path2+'reg_test.csv','r',encoding='utf8')
test= pd.read_csv(f,header=0,sep=',')
test.rename(columns={'c95.1':'c95','c95':'c_95'},inplace = True)
print(test.shape)
print(test)

line =Line()
week=list(test['week'])
c=list(test['c'])
c_95=list(test['c_95'])
c95=list(test['c95'])
```

```python
line = (
    Line()
    .add_xaxis(week)
    .add_yaxis("", c,label_opts=opts.LabelOpts(is_show=False),symbol='circle',
               linestyle_opts=opts.LineStyleOpts(color="red",width=1.5))
    .add_yaxis("", c_95,label_opts=opts.LabelOpts(is_show=False),symbol_size=0,
               linestyle_opts=opts.LineStyleOpts(color="blue",width=1.5,type_='dotted'))
    .add_yaxis("", c95,label_opts=opts.LabelOpts(is_show=False),symbol =None,
               linestyle_opts=opts.LineStyleOpts(color="blue",width=1.5,type_='dotted'))
                 
)
line.set_global_opts(title_opts=opts.TitleOpts(title="回归系数与置信区间", subtitle="副标题"),
                    legend_opts=opts.LegendOpts(is_show=False),
                    xaxis_opts=opts.AxisOpts(name="week",is_show=True,
                                            name_location='middle',name_gap=30),
                    yaxis_opts=opts.AxisOpts(name="回归系数与置信区间",is_show=True,
                                            name_location='middle',name_gap=60))
line.render_notebook() 
#注：上述图置信区间一个有标记，一个没有标记，仅为演示
```

<font face="微软雅黑" size=3>实战3：综合性绘图，几乎所有方面都被修改了</font> 

```python
# 分地区结果
import pandas as pd
import numpy as np
from pyecharts.charts import Line
import pyecharts.options as opts

path="D:/python/郭峰Python讲义/数据与结果/07数据可视化/"
f=open(path+'qt4_mer_cnt.csv','r',encoding='utf8')
data= pd.read_csv(f,header=0,sep=',')

week= list(data['week'])
week=[ str(int(w)) for w in week]
wuhan=list(data['prov1'])
hubei=list(data['prov2'])
jjj=list(data['prov3'])
others=list(data['others'])

line =(
    Line()
   .add_xaxis(week)
   .add_yaxis("武汉",wuhan,is_smooth=True,
              label_opts=opts.LabelOpts(is_show=False),symbol='rect',color="#0025FF",symbol_size=7,
             linestyle_opts=opts.LineStyleOpts(width =2,color = ' #171717',type_='solid'))
    .add_yaxis("湖北",hubei,is_smooth=False,
            label_opts=opts.LabelOpts(is_show=False),symbol='circle',color="#FF1C00",symbol_size=7,
             linestyle_opts=opts.LineStyleOpts(width =2,color = '#06DF06',type_='solid'))
    
    .add_yaxis("京津冀",jjj,is_smooth=False,
            label_opts=opts.LabelOpts(is_show=False),symbol='triangle',color="#06DF06",symbol_size=7,
             linestyle_opts=opts.LineStyleOpts(width =2,color = '#FF1C00',type_='solid'))
            
    .add_yaxis("其他地区",others,is_smooth=False,
               label_opts=opts.LabelOpts(is_show=False),symbol='diamond',color="#171717",symbol_size=7,
              linestyle_opts=opts.LineStyleOpts(width =2,color = '#0025FF',type_='solid'),
              markline_opts=opts.MarkLineOpts(
                   data=[opts.MarkLineItem(name="分割线",x=4)],symbol ='none',
                   label_opts=opts.LabelOpts(is_show=True,position='top'), 
                   linestyle_opts=opts.LineStyleOpts(width =1.5,color = '#000000',type_='dotted')))
    .set_global_opts(title_opts=opts.TitleOpts(title="结果", subtitle=None,pos_top=True,pos_right="60%",
                                              title_textstyle_opts=opts.TextStyleOpts(font_size=20)),
                    legend_opts=opts.LegendOpts(is_show=True,pos_bottom='12%',pos_left='10%',orient='vertical'),
                    xaxis_opts=opts.AxisOpts(name="2020年1月20日后第几周",is_show=True,
                                            axistick_opts=opts.AxisTickOpts(is_inside=True,is_align_with_label=True),
                                            name_location='middle',name_gap=30,
                                            name_textstyle_opts=opts.TextStyleOpts(font_size=20,font_style='normal',font_weight='bolder'),
                                            axislabel_opts=opts.LabelOpts(font_size=15),
                                            axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(width =2,color = '#000000'))),
                    yaxis_opts=opts.AxisOpts(name="分析结果",is_show=True,
                                            axistick_opts=opts.AxisTickOpts(is_inside=True),
                                            name_location='middle',name_gap=40,split_number=6,min_=0.2,max_=0.9,
                                            name_textstyle_opts=opts.TextStyleOpts(font_size=20,font_style='normal',font_weight='bolder'),
                                            axislabel_opts=opts.LabelOpts(font_size=15),
                                            axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(width =2,color = '#000000'))))
)
line.render(path+"shock_diff_prov_cnt.html")    # 生成HTML 文件
line.render_notebook() 
```
